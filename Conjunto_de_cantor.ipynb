```python
def cambiar_decimales_de_base_n_a_base_10(base_n, lista_de_decimales):
  decimales = 0
  for i in range(len(lista_de_decimales)):
    decimales += lista_de_decimales[i]*(base_n**(-i-1))
  return decimales
```

```python
cambiar_decimales_de_base_n_a_base_10(3, [1,0,1])
```

```python
conjunto_cantor = [[0],[2]]
```

```python
def agregar_precision_cantor(n_trecimales):
  for j in range(n_trecimales):
    long = len(conjunto_cantor)
    for i in range(long):
      nuevo_0 = conjunto_cantor[i].copy()
      nuevo_0.append(0)
      nuevo_2 = conjunto_cantor[i].copy()
      nuevo_2.append(2)
      conjunto_cantor.append(nuevo_0)
      conjunto_cantor.append(nuevo_2)
```

```python
conjunto_cantor = [[0],[2]]
agregar_precision_cantor(10)
```

```python
lista_decimales=[]
lista_alturas = []
tamanios = []
for i in range(len(conjunto_cantor)):
  x = conjunto_cantor[i]
  if (len(x)>3):
    decimal = cambiar_decimales_de_base_n_a_base_10(3,x)
    lista_decimales.append(decimal)
    lista_alturas.append(1/len(x))
    tamanios.append(1/(10*len(x)))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(1,figsize=(100, 50))
plt.scatter(lista_decimales, lista_alturas, s=tamanios, alpha=0.5)
plt.show()
```

```python
conjunto_cantor = [[0],[2]]
agregar_precision_cantor(5)
```

```python
cantor_x = []
cantor_y = []
tamanios = []
for i in range(len(conjunto_cantor)):
  x = cambiar_decimales_de_base_n_a_base_10(3,conjunto_cantor[i])
  for j in range(len(conjunto_cantor)):
    y=cambiar_decimales_de_base_n_a_base_10(3,conjunto_cantor[j])
    cantor_x.append(x)
    cantor_y.append(y)
    tamanios.append(1/(10*max(len(conjunto_cantor[i]),len(conjunto_cantor[j]))))
```

```python
fig = plt.figure(1,figsize=(10, 10))
plt.scatter(cantor_x, cantor_y,s=tamanios, alpha=0.5)
plt.show()
```
